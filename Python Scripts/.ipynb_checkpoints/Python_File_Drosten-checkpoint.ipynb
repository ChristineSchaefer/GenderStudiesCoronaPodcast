```python
# myroot[5].attrib.get("sprecher") == 'Korinna Hennig'


import numpy as np
import pandas as pd 
from xml.dom import minidom
import xml.etree.ElementTree as ET
from nltk.tokenize import sent_tokenize
import nltk
import re
nltk.download('punkt')
import os


doc_list = []
fileNameList = []
questionList = []


```

```python


# Diese Funktion speichert in einer ihr übergebenen Liste entweder DROSTEN oder CIESEK, 
#je nachdem wer befragt wird, und gibt diese Liste  anschließend zurück

def whichDoc(xml_path, liste):
    
    #Episode
    #mydoc = minidom.parse(xml_path)

    #episode = mydoc.getElementsByTagName('folge')[0].firstChild.data

   

    tree = ET.parse(xml_path)
    root = tree.getroot()
    for child in root.iter():
        if (child.find("personen")):

            # Iteriert über die einzelnen childnodes von "personen"
            for p in child.find("personen"):

                if "CHRISTIAN DROSTEN" in p.text:
                    liste.append("CD")
                elif "SANDRA CIESEK" in p.text:
                    liste.append("SZ ")
                else:
                    continue

    return liste




def getQuestions(xml_file):
     
   
    sätze_in_liste = []    
    mydoc = minidom.parse(xml_file)    
    
    # Wähle alle "text"- Knoten aus
    texte = mydoc.getElementsByTagName('text')
    
    
    # Iteriere über die einzelnen text-Elemente
    for text in texte:
        if (text.attributes['sprecher'].value == "Korinna Hennig" or text.attributes['sprecher'].value == "Anja Martini"):
            sätze_in_liste.append(text.firstChild.data.strip()) # Strip-Methode entfernt Leerzeichen und \n am Anfang und
            #Ende jedes Satzes
    
    # In der Liste sind mehrere Sätze pro Listenelement enthalten. Diese werden von der Regex-Regel in filterNurFragen
    # nicht korrekt verarbeitet
    text = " ".join(sätze_in_liste)
    
    
    filtered_questions = filterNurFragen(text)
    
    return filtered_questions
   
            
    
   
# Übergeben wird der Text der Sprecherin Korinna Hennig oder Anja Martini und nur ihre Fragen werden zurückgegeben
def filterNurFragen(text):
    
    # Der eingegebene Text wurde nochmal nach jedem einzelnen Satz getrennt
    text = sent_tokenize(text) 
    print(text)
    
   
    
    intern_list = []
   
    for sentence in text:

        #Where ^[A-Z] looks for a Capital at start
        # '[^?!.]*' means everything in between start and end is ok except things containing ? or ! or .
        # [?.!]$ must end with ? or ! or .
        
        if re.match('^[A-Z][\w\s]+[?]$', sentence):
            intern_list.append(sentence)
            
    fragenAlsText =  " ".join(intern_list)
    print(fragenAlsText)
    
    return fragenAlsText


```

# Nachfolgende Zeile generiert die Informationen für das Datenblatt

```python
for dirname, _, filenames in os.walk('Datensatz/'):
    
    
    # Iteriere über alle einzelnen XML Files
    for filename in filenames:
        
       


        # Speichere die Namen der einzelnen Dateien in eine Liste
        fileNameList.append(filename)
        
        
        # f steht für jeweils eine XML-Datei
        
        f = os.path.join(dirname, filename)

        
        # doc_list enthölt den jeweiligen Arzt / die Ärztin für jede XML_File 
        doc_list = whichDoc(f,doc_list)
        
        # questionList ist eine Liste, deren einzelne Listenelemente selbst 
        # aus einer Liste an Fragen pro jede XML-file bestehen.
        questionList.append(getQuestions(f))
    

          

       

    
```

## Erstelle Datentabelle

```python
import pandas as pd

pd.set_option('display.max_rows', None)
  


# intialise data of lists.
data = {'Filename':fileNameList,
        'Questions': questionList,
        'Arzt' :  doc_list      }
  
# Create DataFrame

df = pd.DataFrame(data)

```
